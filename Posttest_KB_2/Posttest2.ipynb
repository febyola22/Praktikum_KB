***1. Menampilkan rata-rata, standar deviasi, nilai minimum, nilai maksimum, Q1, Q2, dan Q3 dari attribute bertipe angka***

```python
import numpy as np
import pandas as pd

ds = pd.read_csv("Data.csv");

atribut_angka = ds.select_dtypes(include=['number'])

statistics = atribut_angka.describe()

# Tambahkan perhitungan kuartil Q1, Q2 (median), dan Q3
statistics.loc['Q1'] = atribut_angka.quantile(0.25)
statistics.loc['Q2'] = atribut_angka.median()
statistics.loc['Q3'] = atribut_angka.quantile(0.75)

print(statistics)
```

***3. Menampilkan informasi jumlah record, jumlah attribute, nama attribute, jumlah attribute dengan tipe angka. Masukkan informasi tersebut dalam dictionary dengan key sesuai nama informasinya. Tampilkan isi dari dictionary tersebut sebagus mungkin***

```python
import pandas as pd
ds = pd.read_csv("Data.csv")

atribut_angka = ds.select_dtypes(include=['number'])
jumlah_records = len(atribut_angka)
jumlah_attributes = len(ds.columns)
nama_attributes = ds.columns.tolist()
atribut_angka = ds.select_dtypes(include=['number'])
jumlah_atribut_angka = len(atribut_angka.columns)

# Membuat dictionary
dataset_info = {
    'Jumlah Record': jumlah_records,
    'Jumlah Attribute': jumlah_attributes,
    'Nama Attribute': nama_attributes,
    'Jumlah Attribute dengan Tipe Angka': jumlah_atribut_angka
}

# Menampilkan Isi Dictionary
print("\n=================================")
print("Informasi Dataset:")
print("=================================")
print("-" * 50)
for key, value in dataset_info.items():
    if key == 'Jumlah Attribute dengan Tipe Angka':
        print(f"{key:<40}: {value}")
    else:
        print(f"{key:<40}: {value}")
print("-" * 50)
```
